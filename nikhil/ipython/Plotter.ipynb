```python
col_names = cr_dfs[0].columns

def show_plot_(row, col):
#     cr_dfs[0].values
    corr = []
    for df in cr_dfs:
        corr.append(df.values[row][col])


    figure()
    plot(scales[0:len(corr)], corr, 'r')
    xlabel(col_names[row])
    ylabel(col_names[col])
    # title('title')
    show()
    
show_plot_(0,0)
show_plot_(12,19)
```

```python
def get_dat(row, col):
#     cr_dfs[0].values
    corr = []
    for df in cr_dfs:
        corr.append(df.values[row][col])
    return corr

    
fig = plt.figure(figsize=(12, 12))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05) 

rows = 12
cols = 12
for i in range(rows):
    for j in range(cols):
        ax = fig.add_subplot(10, cols, (i*cols) + j , xticks=[], yticks=[])
        corr = get_dat(i,j)
        ax.plot(scales[0:len(corr)], corr)
        
```
